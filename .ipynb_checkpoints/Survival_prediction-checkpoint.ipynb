```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
train=pd.read_csv('titanic_train.csv')
```

```python
test=pd.read_csv('titanic_test.csv')
```

```python
train.isnull()
```

```python
sns.heatmap(train.isnull(),yticklabels=False, cbar=False, cmap='viridis')
```

```python
sns.boxplot(x='Pclass', y='Age', data=train)
```

```python
def compute_age(cols):
    Age=cols.iloc[0]
    Pclass=cols.iloc[1]

    if pd.isnull(Age):
        if Pclass==1:
            return 37
        elif Pclass==2:
            return 29
        else:
            return 24
    else:
        return Age
```

```python
train['Age']=train[['Age','Pclass']].apply(compute_age,axis=1)
```

```python
sns.heatmap(train.isnull(),yticklabels=False, cbar=False, cmap='viridis')
```

```python
train.drop('Cabin',axis=1,inplace=True)
```

```python
train.info()
```

```python
train.dropna(inplace=True)
```

```python
train.info()
```

```python
sex=pd.get_dummies(train['Sex'],drop_first=True).astype(int)#we did drop_first to remove female column so that one colujn is not a perfect indicator of another
sex.head()
```

```python
embarked=pd.get_dummies(train['Embarked'],drop_first=True).astype(int)#both columns aren't perfect predictor of each other as there were total thrre arguments
embarked.head()
```

```python
train = train.reset_index(drop=True)
train = pd.concat([train, sex, embarked], axis=1)
```

```python
train=train.drop(['Name','Ticket','Embarked','Sex'],axis=1)
```

```python
train.drop('PassengerId',axis=1, inplace=True)
```

```python
train = train.loc[:, ~train.columns.duplicated()]
```

```python
train.head()
```

```python
train.dropna(inplace=True,axis=0)
```

```python
train.columns
```

```python
X=train[['Pclass', 'Age', 'SibSp', 'Parch', 'Fare', 'male', 'Q', 'S']]
y=train['Survived']
```

```python
#X=train[['Pclass', 'Age', 'SibSp', 'Parch', 'Fare', 'male', 'Q', 'S']]
#y=train['Survived']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lm=LogisticRegression(max_iter=1000)
```

```python
lm.fit(X_train,y_train)
```

```python
predictions=lm.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,predictions))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,predictions)
```
